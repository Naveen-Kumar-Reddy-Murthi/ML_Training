```python
import numpy as np
import pandas as ps

raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
        'age': [42, np.nan, 36, 24, 73], 
        'sex': ['m', np.nan, 'f', 'm', 'f'], 
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = ps.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
df_no_missing = df.dropna()
df_no_missing = df.dropna(how='all')
df_no_missing
```

```python
# Adding new column with default entries to existing data frame

df['location'] = "Hyderabad"
df
```

```python
df['location1']= np.nan
df
```

```python
# remove a column if values of a column are NaN
df.dropna(axis=1, how='all')
df
```

```python
dfnew = df.dropna(thresh=5)
#dfnew
```

```python
# Fill missing column value with mean value of that column
df["preTestScore"].fillna(df["preTestScore"].mean(),inplace=True)
df
```

```python
#Fill missing column value with group average
df["postTestScore"].fillna(df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)  
df
```

```python
import seaborn as sb
import matplotlib.pyplot as mpl
import pandas as pd
%matplotlib inline

url = "D:/data/iris.csv"

data = pd.read_csv(url)
data.head()
#statistical summary
# pd.set_option('precision',2)
data.describe()

mpl.style.use('ggplot')
sb.set(style="white", color_codes=True)
data["species"].value_counts()
```

```python
data.groupby("species").describe().T
```

```python
data.plot(kind="scatter", x="sepal_length", y="sepal_width")
```

```python
#same chart with seaborn
sb.jointplot(x="sepal_length", y="sepal_width",data=data)
```

```python
sb.FacetGrid(data,hue="species", height=5)\
.map(mpl.scatter,"sepal_length", "sepal_width")\
.add_legend()
```

```python
sb.boxplot(x="species", y="petal_length", data=data)
```

```python
sb.violinplot(x="species", y="petal_length", data=data)
```

```python
sb.pairplot(data,hue="species")
```

```python
data.boxplot(by="species", figsize=(25,20))
```
